```python

from sklearn.datasets import load_boston
import pandas as pd


```

```python
boston=load_boston()

x=boston.data #features
y=boston.target #labels
```

```python
#save features
col=boston.feature_names 
print(col)
```

```python
#original dataset

boston_df = pd.DataFrame(x)
#fix columns name
boston_df.columns=col
```

```python
# create 2 copies

boston_df_o1 =pd.DataFrame(x) #z_score
boston_df_o1.columns=col
boston_df_o2= pd.DataFrame(x) #IQR_score
boston_df_o2.columns=col
```

```python
# check if there is missing values
# as it is same as the real number of rows, it is complete
boston_df.describe()
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# use DIS feature to detect the outlier
sns.boxplot(x=boston_df['DIS'])
```

```python
#scatter

fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(boston_df['INDUS'], boston_df['TAX'] )
ax.set_xlabel('proporation of non-retail business acres per town')
ax.set_ylabel('Full_value property-tax rate per $10,000')
plt.show()
```

```python
#z_score
from scipy import stats #z_score
z = np.abs(stats.zscore(boston_df)) ##calculates the Z-score
print(z)
```

```python
Q1 = boston_df_o1.quantile(.25)
Q3 = boston_df_o2.quantile(.75)
IQR = Q3 - Q1
print(IQR)
#for each feature
```

```python
boston_df.shape
```

```python
##removing outlier based on z-score
boston_df_o1 = boston_df_o1[(z < 3).all(axis=1)]
##any point is greater than 3 ,will be removed
```

```python
#100 columns are removed
boston_df_o1.shape
```

```python
#removing the outlier
boston_df_o2= boston_df_o2[~((boston_df_o2 < (Q1 - 1.5 * IQR))| (boston_df_o2 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print(boston_df_o2.shape)
print(boston_df.shape)
print(y.shape)
```
